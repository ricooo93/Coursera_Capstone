```python
from bs4 import BeautifulSoup
from collections import defaultdict
import  urllib.request
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
res = urllib.request.urlopen(url)
html = res.read().decode('utf-8')
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
data = defaultdict(list)
table = soup.find_all('table')[0]
for tr in table.find_all('tr')[1:]:
    postcode, borough, neigh = tr.find_all('td')
    postcode = postcode.string
    borough = borough.string
    if neigh.a:
        neigh = neigh.a.string
    else:
        neigh = neigh.string
    neigh = neigh.strip()
    if borough != 'Not assigned':
        if not data[postcode]:
            data[postcode].append(borough)
        if neigh == 'Not assigned':
            data[postcode].append(borough)
        else:
            data[postcode].append(neigh)
for key, value in data.items():
    data[key] = [value[0]] + [','.join(value[1:])]
#print(data)
```

```python
import pandas as pd
```

```python
converted = {'Postcode':list(data.keys()),'Borough':[v[0] for v in data.values()],'Neighbourhood':[v[1] for v in data.values()]}
df = pd.DataFrame(converted)
df = df[['Postcode','Borough','Neighbourhood']]
df
```

```python
df.shape
```

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')
df_geo.head()
```

```python
posts = df['Postcode'].values
```

```python
latitudes = [df_geo.loc[df_geo['Postal Code']==p].Latitude.values[0] for p in posts]
longitudes = [df_geo.loc[df_geo['Postal Code']==p].Longitude.values[0] for p in posts]
```

```python
df['Latitude'] = latitudes
df['Longitudes'] = longitudes
df
```
